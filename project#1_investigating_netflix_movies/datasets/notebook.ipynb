```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix_df = pd.read_csv("./netflix_data.csv")
```

```python
no_tv = netflix_df["type"] != "TV Show" # initalise boolean mask by setting condition 
netflix_subset = netflix_df[no_tv] # apply boolean mask to filter df
```

```python
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]
```

```python
under_one_hour = netflix_movies["duration"] < 60 # initalise boolean mask by setting condition
short_movies = netflix_movies[under_one_hour] # apply boolean mask to filter df
```

```python
colors = []
for index, row in netflix_movies.iterrows():
    if row["genre"] == "Children":
        colors.append("yellow")
    elif row["genre"] == "Documentaries":
        colors.append("green")
    elif row["genre"] == "Stand-up":
        colors.append("yellow")
    else:
        colors.append("black")
```

```python
figure = plt.figure(figsize=(12,8))
plt.scatter(
    netflix_movies["release_year"], 
    netflix_movies["duration"], 
    c = colors)
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
plt.show()
```
