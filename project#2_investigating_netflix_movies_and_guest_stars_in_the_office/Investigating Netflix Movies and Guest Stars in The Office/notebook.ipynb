```python
# importing modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load csv to dataframe
office_df = pd.read_csv("datasets/office_episodes.csv")
print(office_df.columns)
```

```python
# creating an array for colors based on scaled ratings to be later applied to the scatter plot
colors = []
for index, row in office_df.iterrows():
    if row["scaled_ratings"] < 0.25:
        colors.append("red")
    elif row["scaled_ratings"] >= 0.25 and row["scaled_ratings"] < 0.5:
        colors.append("orange")
    elif row["scaled_ratings"] >=0.5 and row["scaled_ratings"] < 0.75:
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")
```

```python
# creating array with the sizes based on if has_guests to be later applied to scatter plot
sizes = []
for index, row in office_df.iterrows():
    if row["has_guests"] == True:
        sizes.append(250)
    else:
        sizes.append(25)
```

```python
# plotting scatter plot
fig = plt.figure()
plt.rcParams['figure.figsize'] = [11, 7]
plt.scatter(
    office_df["episode_number"], 
    office_df["viewership_mil"], 
    c=colors, 
    s=sizes)
plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.show()
```
